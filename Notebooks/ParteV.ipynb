# Gráficos 2D

```python
# Importação da biblioteca
import matplotlib.pyplot as plt
```

## Gráficos simples

```python
# Gerando a um gráfico simples
fig = plt.figure() # Criação de uma figura
ax = fig.add_subplot() # Adicionando um subplot à figura

# Dados
x = [0, 1, 2, 3, 4]
y = [1, 2, 3, 4, 5]

# Construção do gráfico
ax.plot(x, y)

# Mostra o gráfico
plt.show()
```

```python
# Adicionando título, legenda e nome aos eixos
fig = plt.figure() # Criação de uma figura
fig.suptitle("Título da figura") # Título da figura

ax = fig.add_subplot() # Adicionando um subplot à figura
ax.set(xlabel='x', ylabel='y', title='Título do gráfico') # Colocando o nome dos eixos e título

# Dados
x = [0, 1, 2, 3, 4]
y = [1, 2, 3, 4, 5]

# Construção do gráfico
ax.plot(x, y, label='Legenda')
ax.legend() # Mostrar e configurar a legenda

# Mostra o gráfico
plt.show()
```

```python
# Gráfico com dados diferentes
import numpy as np

fig, ax = plt.subplots() # Cria a figura e o subplot

# Dados para o gráfico
x = np.linspace(0, 2*np.pi, 50)
seno = np.sin(x) # Função seno
cosseno = np.cos(x) # Função cosseno

ax.plot(x, seno, label='Seno') # Plota a função seno
ax.plot(x, cosseno, label='Cosseno') # Plota a função cosseno

# Outra forma de nomear os eixos e o título de um subplot
ax.set_xlabel("x", fontsize=16)
ax.set_ylabel("Amplitude", fontsize=16)
ax.set_title("Seno e Cosseno", fontsize=16)

ax.legend() # Mostrar e configurar a legenda

# Mostra o gráfico
plt.show()
```

## Figura com múltiplos subplots

```python
# Figura com multiplos subplots
fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(10, 5)) # Criando a figura com dois subplots

ax1.plot(x, seno, label='Seno') # Plota a função seno
ax1.set_xlabel('x', fontsize=20) # Nomeação do eixo x
ax1.set_ylabel('Amplitude', fontsize=20) # Nomeação do eixo y
ax1.set_title('Função Seno', fontsize=20) # Título do subplot

ax2.plot(x, cosseno, label='Cosseno') # Plota a função cosseno
ax2.set_xlabel('x', fontsize=20) # Nomeação do eixo x
ax2.set_ylabel('Amplitude', fontsize=20) # Nomeação do eixo y
ax2.set_title('Função Cosseno', fontsize=20) # Título do subplot

# Ajustando a distribuicao dos graficos na figura
plt.tight_layout()

# Mostra os gráficos
plt.show()
```

```python
# Diferentes tipos de gráficos
# Dados
x = np.linspace(0, 10, 10)
y = np.linspace(0, 100, 10)
z = np.linspace(0, 20, 10)

# Criando a figura
fig = plt.figure(figsize=(14, 12))

# Adicionando os eixos
ax1 = fig.add_subplot(321)
ax2 = fig.add_subplot(322)
ax3 = fig.add_subplot(323)
ax4 = fig.add_subplot(324)
ax5 = fig.add_subplot(325)
ax6 = fig.add_subplot(326)

# Construindo os graficos
ax1.bar(x, y, color='red') # Gráfico de barras
ax2.barh(x, y, color='orange') # Gráfico de barras horizontais
ax3.axhline(5) # Linha horizontal
ax4.axvline(50) # Linha vertical
ax5.plot(x, y, linestyle='-.', color='m')
ax6.scatter(x, y, marker='^', color='y') # Gráfico de pontos

# Ajustando a distribuicao dos graficos na figura
plt.tight_layout()

# Mostrando os graficos
plt.show()
```

## Histograma

```python
# Criando um histograma 
# Dados
x = np.random.randn(10000) # Amostras de uma distribuição normal padrão

# Gerando o histograma
'''
fig,ax = plt.subplots()
ax.hist(x, bins=20)
ax.set_title("Um Histograma")
'''

plt.hist(x, bins=20)
plt.title("Um Histograma")

# Mostra o gráfico
plt.show()
```

```python
# Histograma normalizado - PDF
# Dados
y = np.random.random(10000) # Amostras de uma distribuição uniforme

# Gerando o histograma
plt.hist(y, bins=20, density=True)
plt.title("Um Histograma Normalizado")

# Mostra o gráfico
plt.show()
```

```python
# Histograma - CDF
# Gerando o histograma
plt.hist(y, bins=50, density=True, cumulative=True)
plt.title("Um Histograma CDF")

# Mostra o gráfico
plt.show()
```

# Gráficos 3D

```python
from mpl_toolkits.mplot3d import Axes3D # Necessário para construção dos gráficos 3D
import matplotlib.pyplot as plt
import numpy as np

# Dados
x = np.linspace(-100, 100, 50)
y = np.linspace(-100, 100, 50)
X, Y = np.meshgrid(x, y) # Cria uma grade de valores 
Z = Y**2/4 - X**2/16

# Construção do gráfico
fig = plt.figure(figsize=(7, 5))
ax = fig.add_subplot(projection='3d')
ax.scatter(X, Y, Z)
ax.set(xlabel='x', ylabel='y', zlabel='z', title='Meu primeiro gráfico 3D')

# Mostra o gráfico
plt.show()
```

```python
# Superfície
fig = plt.figure(figsize=(7, 5))
ax = fig.add_subplot(projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis') # Plota a superfície
ax.set(xlabel='x', ylabel='y', zlabel='z', title='Minha primeira superfície')

# Show the plot
plt.show()
```

```python
# Superfície de contorno
fig = plt.figure()
ax = fig.add_subplot()

C = ax.contour(X, Y, Z, cmap='inferno') # Plota a superfície de contorno

ax.set(xlabel='x', ylabel='y', title='Meu primeiro contorno')
ax.clabel(C) # Legenda da superfície de contorno
plt.show()
```
